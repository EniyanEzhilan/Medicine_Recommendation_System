## Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X = df.drop(columns="Disease", axis = 1)
y = df["Disease"]
```

```python
Label_obj = LabelEncoder()
Label_obj.fit(y)
Y = Label_obj.transform(y)
Y
```

```python
X
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state=2, test_size=0.2)
```

## Let us train all models to know which is the best model

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np


# Create a dictionary to store models
models = {
    'SVC': SVC(kernel='linear'),
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    'GradientBoosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'KNeighbors': KNeighborsClassifier(n_neighbors=5),
    'MultinomialNB': MultinomialNB()
}
```

```python
print(list(models))
```

```python
for key,model in models.items():
    #let us fit in the first model
    model.fit(X_train,Y_train)
    #Now let us train the model
    predicted_values = model.predict(X_test)

    #let us find the accuracy
    acc_score = accuracy_score(Y_test,predicted_values)

    #lets do the same for confusion matrix

    con_matrix = confusion_matrix(Y_test,predicted_values)
    
    print("model name:", key)
    print("Accuracy_Score: ", acc_score)
    print("Confusion_Matrix: ", con_matrix)
    
```

## Choosing the best model

```python
#I'm choosing SVM out of all this

svc_obj = SVC(kernel = 'linear')
svc_obj.fit(X_train,Y_train)
pred = svc_obj.predict(X_test)
#pred
```
